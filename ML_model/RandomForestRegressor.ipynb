```python
import numpy as np
import pandas as pd

df = pd.read_csv('/gdrive/My Drive/Colab Notebooks/dust_project/DataSet.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(20,10))
sns.heatmap(data=df.corr(method='spearman'), annot=True,fmt='.2f', linewidths=.5, cmap='Blues')
```

```python
x = df[['Temp', 'Prec', 'WS', 'Humi', 'Pres', 'cbwd']]

y = df[['PM10']]
```

```python
# minmaxscaler
# 0,1 사이로
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler_col = ['Temp', 'Prec', 'WS', 'Humi', 'Pres', 'cbwd', 'PM10']
df_scaled = scaler.fit_transform(df[scaler_col])
df_scaled = pd.DataFrame(df_scaled)
df_scaled.columns = scaler_col

df_scaled.head()
```

```python
feature_cols = ['Temp', 'Prec', 'WS', 'Humi', 'Pres', 'cbwd']
label_cols = ['PM10']
x = df[feature_cols]
y = df[label_cols]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor # 회귀트리(모델)
from sklearn.metrics import mean_squared_error, r2_score

rmses = []

for depth in range(1,25):
  model = RandomForestRegressor(min_samples_split=3, max_depth = depth)
  model.fit(X = x_train, y = y_train)
  rmse = mean_squared_error(y_test, model.predict(x_test))**0.5
  print('depth = {0:2d} rmse: : {1:.3f}'.format(depth, rmse))
  rmses.append(rmse)


rmses = pd.DataFrame({'RMSE':rmses})
rmses.plot()
```

```python
print(f"RFR 훈련 세트 점수: {model.score(x_train, y_train)} \nRFR 테스트 세트 점수 : {model.score(x_test, y_test)}")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_test, model.predict(x_test))
mse = mean_squared_error(y_test, model.predict(x_test))
print('===== 결과 =====')
print('score : %.3f' % model.score(x_test, y_test))
print('RMSE: %.3f' % mse**0.5)
```
